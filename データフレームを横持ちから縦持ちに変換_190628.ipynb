### カラム名が「数字のみ」のカラムのみに特定文字列を追加する

```python
col_list = df.columns.tolist()

for col in col_list:
    if str(col).isdecimal() is True:
        df= df.rename(columns={col:str(col)+'_回数'})
    else:
        pass
```

# 横持ちから縦持ちに変換

```python
df1 = df[['ID', '2015以前', '2016_回数', '2017_回数', '2018_回数']]

unpivot_df = pd.melt(df1, id_vars=['ID'], var_name='年', value_name='回数')

unpivot_df
```

## 縦持ちから横持ちに変換

```python
pd.pivot_table(data=unpivot_df, index='ID',columns='年', values='回数', aggfunc=sum)
```
