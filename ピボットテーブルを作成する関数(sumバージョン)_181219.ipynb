```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_rows = 1000
pd.options.display.max_columns = 1000
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
def create_sum_pivot_table(df, index, columns, values):
    pivot_df = pd.pivot_table(
    df,
    index=index,
    columns=columns,
    values=values,
    aggfunc='sum',
    fill_value=0
    )
    return pivot_df.head()
```

```python
create_sum_pivot_table(df, 'week', 'name', 'soldout')
```
