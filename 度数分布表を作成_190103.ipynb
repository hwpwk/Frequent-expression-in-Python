```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('all_text_wakatigaki.csv',encoding='utf-8')
df.head()
```

```python
cnt_df = df['original'].value_counts().reset_index().rename(columns={'index':'word', 'original':'度数'})
display(cnt_df.head(), cnt_df.tail())
```

## 度数分布表の作成

```python
cnt_df['累積度数'] = cnt_df['度数'].cumsum()

cnt_df['相対度数'] = round((cnt_df['度数'] / sum(cnt_df['度数'])), 2)

cnt_df['累積相対度数'] = round((cnt_df['累積度数'] / sum(cnt_df['度数'])), 2)

display(cnt_df.head(), cnt_df.tail())
```
