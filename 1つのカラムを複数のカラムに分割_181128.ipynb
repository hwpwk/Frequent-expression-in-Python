```python
# データ分析用のライブラリ
import pandas as pd
import numpy as np
```

#### データ読み込み時にカラムを絞る
usecols=['xx', 'yy']を引数に与える

```python
squeeze_df = pd.read_csv('all_text_wakatigaki.csv', usecols=['内容', 'original', 'type'] ,encoding='utf-8')
squeeze_df.head()
```

#### 複数列の文字列同士を結合して新たな列を生成
「+」演算子を使う

```python
squeeze_df['original_type'] = squeeze_df['original'] + ' _ ' + squeeze_df['type']
squeeze_df.head()
```

## 文字列要素を持つカラムを複数のカラムに分割
特定の列にstr.split()を使うと分割されたデータフレームができるのでこれを分割前のデータフレームに連結し、元のカラムを除外する

```python
df = pd.concat([squeeze_df, squeeze_df['original_type'].str.split('_', expand=True)], axis=1).drop('original_type', axis=1)
df = df.rename(columns={0:'word', 1:'type1', 2:'type2'})
df
```
