## timedelta64[ns]型の値から日付部分のみ抽出してint型に変換する

```python
# まずはstr型に変換
df['期間'] = df['期間'].astype(str)

df.head(1)
```

```python
# 文字列「 days」でsplitして0番目のみを抽出
df['期間_修正'] = df['期間'].map(lambda x: x.split(' days')[0]).astype(int)

display(df, df.dtypes)
```
