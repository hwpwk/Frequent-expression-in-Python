```python
import pandas as pd
import numpy as np
import glob
```

### カレントディレクトリに格納されている全xlsxファイル名の抽出

```python
files = glob.glob('*xlsx')

files
```

# 読み込んでいくデータフレームを辞書型orリスト型で格納

```python
# 辞書を使った場合
df_dict = {}

for i, file in enumerate(files):
    df_dict[i] = pd.read_excel(file)

    
# 上記を辞書型内包表記にした場合
df_dict2 = {i: pd.read_excel(file) for i, file in enumerate(files)}
```

```python
# リスト型でも格納加納
df_list = [pd.read_excel(file) for file in files]
```

### 辞書&リストにデータフレームが格納されているかアクセスする

```python
for i in range(len(df_dict)):
    display(df_dict[i].head(2), df_dict2[i].head(2), df_list[i].head(2))
```
