```python
import pandas as pd
```

```python
df = pd.read_excel('横持ちデータ 201911.xlsx')

df
```

# 横持ちデータから縦持ちデータへの変換


id_vars：基本軸となるカラムを指定

var_name：変数(Variable)となるカラムに名前を新規でつける

value_name：値(Value)となるカラムに名前を新規でつける
    
参考：https://qiita.com/ishida330/items/922caa7acb73c1540e28


```python
tate_df = pd.melt(df, id_vars=['科目名'], var_name='企業名', value_name='金額')

tate_df
```

## 縦持ちデータから横持ちデータへの変換

```python
yoko_df = pd.pivot_table(tate_df, index='科目名', columns='企業名', values='金額', aggfunc=sum)

yoko_df 
```

```python
# データフレームの整形 

yoko_df = yoko_df.reset_index()
# ピボットテーブル後index名を削除
yoko_df.columns.name = None

yoko_df
```
