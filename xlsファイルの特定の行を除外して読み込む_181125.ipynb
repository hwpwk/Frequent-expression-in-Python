```python
import pandas as pd
import numpy as np

# xlsファイル読み込み用
import xlrd
import pprint
```

特定の行をスキップ（除外）して読み込む場合、引数skiprowsを使う。

skiprowsに整数を渡すと、ファイルの先頭をその行数分スキップして読み込む。

```python
df = pd.read_excel('utility_charges.xls', encoding='utf-8', skiprows=2)
df
```

```python
# インデックス番号で除外する行を指定することも可能（その場合はリスト型にする)
df2 = pd.read_excel('utility_charges.xls', encoding='utf-8', skiprows=[0,1,3,4])
df2
```

```python
# header= で指定した行から読み込むことも可能
df3 = pd.read_excel('utility_charges.xls', encoding='utf-8', header=2)
df3
```
