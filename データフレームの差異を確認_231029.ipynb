```python
import pandas as pd

data1 = {'会社コード': [1, 2, 3, 4, 5, 6, 7, 8],
         '会社名': ['会社A', '会社B', '会社C', '会社D', '会社E', '会社F', '会社G', '会社H'],
         '科目コード': [101, 102, 103, 104, 105, 106, 107, 108],
         '科目名': ['科目1', '科目2', '科目3', '科目4', '科目5', '科目6', '科目7', '科目8']}

data2 = {'会社コード': [1, 2, 3, 4, 5, 6, 7, 8],
         '会社名': ['会社A', '会社B', '会社C', '会社D', '会社E', '会社F', '会社G', '会社H'],
         '科目コード': [101, 102, 103, 104, 105, 106, 107, 108],
         '科目名': ['科目1', '科目2', '科目3', '科目4', '科目5', '科目6', '科目7', '科目8']}

data3 = {'会社コード': [1, 2, 3, 4, 5, 6, 7, 8],
         '会社名': ['会社A', '会社B', '会社CC', '会社D', '会社E', '会社F', '会社G', '会社HH'],
         '科目コード': [101, 102, 103, 104, 105, 106, 107, 108],
         '科目名': ['科目1', '科目2', '科目3', '科目44', '科目5', '科目66', '科目77', '科目8']}


df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
```

```python
display(df1, df2, df3)
```

```python
dfs_dic = {'df2':df2, 'df3':df3}

# 差分があった場合に差分を確認しやすいよう「index」列を付与(何度も実行するとインデックス列が増加するので注意)
df1 = df1.reset_index()
for name, df in dfs_dic.items():
  # 差分があった場合に差分を確認しやすいよう「index」列を付与
  df = df.reset_index()
  df_merge_diff = pd.merge(
      df1, df, on=df1.columns.tolist(), how='outer', indicator=True
      ).query('_merge!="both"'
      )
  if len(df_merge_diff) == 0:
    print(f'OK!df1と{name}に差異はありません。')
  else:
    diff_cols_list = []
    for col in df1.columns.tolist():
      # 指定列のすべての値が等しくない場合はその列名をリストに格納
      if not (df1[col]==df[col]).all():
        diff_cols_list.append(col)
    print(f'NG!df1と{name}は{diff_cols_list}列の値に差異があります。')
    for diff_col in diff_cols_list:
      print(f'差異が発生した列: {diff_col}')
      df_diff = df_merge_diff[['index', diff_col]]
      # 重複行は「差異なし」なので削除するため重複行のインデックス番号を取得
      duplicates_index = df_diff[df_diff.duplicated(keep=False)].index
      # 重複行を削除
      df_diff = df_diff[['index', diff_col]].drop(duplicates_index)
      # 「index」列毎に値の差異を表示
      for _, df_index_group in df_diff.groupby('index'):
        display(df_index_group)
        print()
      print('-'*100)
```
