```python
import pandas as pd
import numpy as np
import re
```

```python
# 適当なデータフレームを作成

df = pd.DataFrame({
    'products_1':['1999年', '((', '=', '1日', '1%', '10%', '2月', '11月', '水筒1', '水筒2', '水筒1', '水筒2', 'フライパン1', 'フライパン2'],
    'price':['1000', '2000', '1000', '2000', '1000', '2000', '1000', '2000','1000', '2000', '1000', '2000', '2000', '3000']
})

df
```

# 数字&記号が含まれる文字列を正規表現で別の文字列(-1)に変更

```python
df['delete_flag'] = df['products_1'].apply(lambda x: re.sub('\d{1,4}[%月年日]|[()=]', '-1', x))
df
```

```python
'''delete_flagカラムで「-1」が含まれているレコード「以外」を抽出'''
# 「以外」は「~(チルダ)」を付ける
df[~(df['delete_flag'].str.contains('-1'))]
```

```python
'''delete_flagカラムで「-1」が含まれているレコードのみを抽出'''
df[df['delete_flag'].str.contains('-1')]
```
