## あるカラムの値が別のカラムの値の中に含まれているかどうかを判定する
参考：https://qiita.com/kishiyama/items/cf3f4dca5ee73dacc551#_reference-0a563087110114a70880

```python
def judge_include_value(df, value, base_col):
    '''
    指定したカラムの値が別のカラムの値の中に含まれているかどうかを判定する関数
    
    Parameters
    ----------
    df : 
        該当データフレーム

    value : str
        含まれているかどうか判定したい値

    base_col : str
        カラム名(このカラムの値の中にvalueが含まれているかどうかを判定)

    Returns
    -------
    type : str
        要素が含まれていれば「1」、含まれていなければ「0」を返す
        
    使い方
    -------
    df['flag'] = [judge_include_number(df, value, 'id_b') for value in df['id'].tolist()]
    
    '''
    
    base_list = df[base_col].tolist()
    
    if value in base_list:
        return '1'
    else:
        return '0'
```

```python
df['flag'] = [judge_include_value(df, value, 'id_b') for value in df['id'].tolist()]
df
```
