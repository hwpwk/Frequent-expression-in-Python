```python
fizzbuzz_list1 = []

for i in range(0, 31):
    if i % 15 == 0:
        fizzbuzz_list1.append('fizzbuzz')
    elif i % 5 == 0:
        fizzbuzz_list1.append('buzz')
    elif i % 3 == 0:
        fizzbuzz_list1.append('fizz')
    else:
        fizzbuzz_list1.append(i)
        
fizzbuzz_list1
```

# リスト内包表記内でfor文-if-elif-else文を使用

```python
fizzbuzz_list2 = ['fizzbuzz' if i % 15 == 0 else 'buzz' if i % 5 == 0 else 'fizz' if i % 3 == 0 else i for i in range(0, 31)]
fizzbuzz_list2
```

### 出力結果が同じであるかを確認

```python
import pandas as pd

df = pd.DataFrame({
    'list1':fizzbuzz_list1,
    'list2':fizzbuzz_list2
})

df.head()
```

```python
'''同じ値なら「0」、違う値なら「1」をつける'''
df['diff_flag'] = df.apply(lambda x: '0' if x['list1'] == x['list2'] else '1', axis=1)

display(df['diff_flag'].value_counts(), df)
```
