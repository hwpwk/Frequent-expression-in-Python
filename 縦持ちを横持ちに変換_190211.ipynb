```python
import pandas as pd
import numpy as np
```

```python
# 適当なデータフレームを2つ作成
# 恣意的に片方のデータフレームのレコード数を増やす
df1 = pd.DataFrame({
    'products':['水筒1', '水筒2', 'フライパン1'],
    'sale_price':[800, 1600, 1800]
})

df2 = pd.DataFrame({
    'customer_id':['1', '2', '3', '4', '5', '6'],
    'products':['水筒1', '水筒2', '水筒1', '水筒2', 'フライパン1', 'フライパン2'],
    'price':[1000, 2000, 1000, 2000, 2000, 3000]
})

display(df1, df2)
```

###  customer_id毎のproductsのsale_priceを紐付け

```python
merge_df = pd.merge(df2, df1, how='left', on='products')
merge_df
```

```python
# aggfunc='mean'を省略しても同じ結果
pivot_df = pd.pivot_table(merge_df, index='customer_id', columns='products', values='sale_price', aggfunc='mean', fill_value=0)
pivot_df
```
