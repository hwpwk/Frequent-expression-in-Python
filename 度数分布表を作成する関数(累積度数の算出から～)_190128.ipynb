## 度数分布表の作成
すでに度数が算出されているデータフレームを使って累積度数から算出したいとき

```python
'''度数分布表を作成する関数(累積度数の算出から～)'''
def create_frequency_distribution_from_cousum(df):
    '''
    df:該当データフレーム
    col:度数、累積度数、相対度数、累積相対度数を算出したいカラム
    
    使い方：create_frequency_distribution_from_cousum(df)
    '''
    #df = df[col].value_counts().reset_index().rename(columns={'index':rename_col, col:'度数'})
    df['累積度数'] = df['度数'].cumsum()
    df['相対度数'] = round((df['度数'] / sum(df['度数'])), 3)
    df['累積相対度数'] = round((df['累積度数'] / sum(df['度数'])), 3)
    return df
```

```python
create_frequency_distribution_from_cousum(cnt_df)
```
