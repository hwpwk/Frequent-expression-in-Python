## np.where関数を使用して条件に応じて値を代入
第一引数に条件、第二引数に第一引数で指定した条件がTrueの時に返す値、Falseの時に返す値を第三引数にする

if then ... else ...のような条件分岐の処理が可能

```python
df['is_match'] = np.where(df['type1'] == df['type2'], 1, 0)

df['is_not_match'] = np.where(df['type1'] != df['type2'], 1, 0)

df
```
