# あるカラムの欠損値を別のカラムの要素で補完する関数

```python
'''あるカラムの欠損値を別のカラムの要素で補完する関数'''
def complement_na_from_oneside(df, col1, col2, new_col):
    '''
    col1:欠損値があれば補完したいカラム
    col2:col1が欠損値のときに欠損していない場合、その要素をcol1に補完するカラム
    new_col:新規に作成するカラム（col1の欠損値が補完された状態）
    '''
    df[new_col] = np.where(df[col1].notnull(), df[col1], np.where(df[col2].notnull(), df[col2], np.nan))
    
    return df
```

```python
new_df = complement_na_from_oneside(df, 'kcal', 'y', 'new_kcal')

display(new_df.isnull().sum(), new_df.head())
```
