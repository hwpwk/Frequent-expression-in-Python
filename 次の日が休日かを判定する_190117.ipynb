```python
import pandas as pd
import numpy as np
```

```python
def basic_check(df):
    '''
    読み込んだデータフレームの
    ・行と列の長さ
    ・各カラムの欠損値の数
    ・各カラムのデータの型
    ・データフレームのレコード先頭から5行、末尾から5行
    を確認する
    '''
    print('行と列の長さ\n{}'.format(df.shape))
    print('-'*50)
    print('各カラムの欠損値の数\n{}'.format(df.isnull().sum()))
    print('-'*50)
    print(df.info())
    display(df.head(), df.tail())
```

```python
df = pd.read_csv('train.csv')
basic_check(df)
```

### datetime型へ変換

datetime型への変換にはpd.to_datetimeメソッドを使い、引数のformatで書式を設定

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%d')
```

### 年月日、曜日を新たなデータとして追加

dayofweekに関しては、0:月曜日、1火曜日、:2:水曜日、3:木曜日、4:金曜日、5:土曜日、6が日曜日を表す

```python
df['day'] = df['datetime'].dt.day
df['month'] = df['datetime'].dt.month
df['dayofweek'] = df['datetime'].dt.dayofweek
df['year'] = df['datetime'].dt.year
```

# 次の日が休日かを判定するカラムを新規追加

金曜、土曜日にフラグを付ける

```python
def judge_nextday_is_holiday(df):
    return np.where((df['dayofweek'] == 4) | (df['dayofweek'] == 5), 1, 0)
```

```python
df['holiday'] = df.apply(judge_nextday_is_holiday, axis=1)
df.head()
```
