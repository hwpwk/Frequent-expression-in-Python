```python
from itertools import product
import pandas as pd
import numpy as np
```

```python
# 年月の組み合わせを生成
dates = [202201, 202202, 202203]

# 2列がイコールになるパターンを含むデータフレームを生成
combinations_all = list(product(dates, repeat=3))
data = [{'計上年月': date1, '出荷年月': date2, '納入年月': date3} for date1, date2, date3 in combinations_all]

# データフレームを作成
df = pd.DataFrame(data)

conditions = [
    (df['計上年月'] < df['出荷年月']) & (df['出荷年月'] < df['納入年月']),
    (df['計上年月'] < df['納入年月']) & (df['納入年月'] < df['出荷年月']),
    (df['出荷年月'] < df['計上年月']) & (df['計上年月'] < df['納入年月']),
    (df['出荷年月'] < df['納入年月']) & (df['納入年月'] < df['計上年月']),
    (df['納入年月'] < df['計上年月']) & (df['計上年月'] < df['出荷年月']),
    (df['納入年月'] < df['出荷年月']) & (df['出荷年月'] < df['計上年月']),
    (df['計上年月'] == df['出荷年月']) & (df['計上年月'] < df['納入年月']),
    (df['計上年月'] == df['出荷年月']) & (df['納入年月'] < df['計上年月']),
    (df['計上年月'] == df['納入年月']) & (df['計上年月'] < df['出荷年月']),
    (df['出荷年月'] == df['納入年月']) & (df['出荷年月'] < df['計上年月']),
    (df['出荷年月'] == df['納入年月']) & (df['計上年月'] < df['出荷年月']),
    (df['計上年月'] == df['納入年月']) & (df['出荷年月'] < df['計上年月']),
    (df['計上年月'] == df['出荷年月']) & (df['計上年月'] == df['納入年月'])
]

choices = [
    '計上<出荷<納入', '計上<納入<出荷', '出荷<計上<納入', '出荷<納入<計上', '納入<計上<出荷', '納入<出荷<計上',
    '計上=出荷<納入', '納入<計上=出荷', '計上=納入<出荷', '出荷=納入<計上', '計上<出荷=納入', '出荷<計上=納入',
    '計上=出荷=納入'
]

df['年月大小比較'] = np.select(conditions, choices, default='その他')

df
```
