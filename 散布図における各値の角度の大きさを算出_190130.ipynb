# 散布図における各値の角度の大きさを算出

2通りある

```python
df['angle'] = np.rad2deg(np.arctan(df['X'] / df['Y']))

df
```

```python
# 180 / πを掛ければラジアンから度に変換できる
df['angle2'] = (np.arctan(df['X'] / df['Y'])) / np.pi *180

df
```
