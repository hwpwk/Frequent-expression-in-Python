## strip関数で文字前後のスペースを削除

```python
# strip関数を使用できるのは文字列型
# よって、astype(str)を付けないとカラムが文字列型以外だとエラーが発生するので注意
u_df = u_df.astype(str).progress_applymap(lambda x: x.strip())
u_df1 = u_df1.astype(str).progress_applymap(lambda x: x.strip())
```

## replace関数で文字間のスペースを削除

```python
u_df = u_df.astype(str).progress_applymap(lambda x: x.replace(' ', ''))
u_df1 = u_df1.astype(str).progress_applymap(lambda x: x.replace(' ', ''))
```

```python
merge_df2 = pd.merge(u_df, u_df1, on='fruits', how='left')
pd.set_option('display.max_rows', 100)
merge_df2
```

```python
merge_df2 = merge_df2.drop_duplicates()
len(merge_df2)
```

```python
merge_df2.isnull().sum()
# strip関数とreplace関数を使うとneologdn.normalize関数を使うのと同様に文字列の前後文字間のスペースを削除することができる
```

## 大文字を小文字に変換

```python
u_df = u_df.astype(str).progress_applymap(lambda x: x.lower())
u_df1 = u_df1.astype(str).progress_applymap(lambda x: x.lower())
```

```python
merge_df3 = pd.merge(u_df, u_df1, on='fruits', how='left')
pd.set_option('display.max_rows', 100)
merge_df3
```

```python
merge_df3 = merge_df3.drop_duplicates()
len(merge_df3)
```

```python
merge_df3.isnull().sum()
# 英字をすべて小文字にすることですべてのキーを紐づけることができた
```
