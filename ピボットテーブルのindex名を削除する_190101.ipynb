```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)

df = pd.DataFrame({
    'id': np.random.randint(low=1, high=30, size=200),
    'cate': np.random.randint(low=1, high=5, size=200),
    'payment': np.random.randint(low=100, high=900, size=200)
})

df.head()
```

```python
pivot_df = df.pivot_table(
    index='id',
    columns='cate',
    values='payment',
    aggfunc='sum',
    fill_value=0
)

display(pivot_df.head(), pivot_df.tail())
```

```python
pivot_df = pivot_df.reset_index()
pivot_df.head()# ピボットテーブルにreset_index関数を適応すると[id]カラムが列に移動して[cate]カラムがindexに残ったままになる
```

## index名を削除

```python
# iindex名'cate'を削除
pivot_df.columns.name = None

display(pivot_df.head(), pivot_df.tail())
```
